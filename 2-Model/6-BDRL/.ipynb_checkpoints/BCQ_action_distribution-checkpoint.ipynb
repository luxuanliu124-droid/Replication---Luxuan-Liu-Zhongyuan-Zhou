```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
df_bcq = pd.read_csv("./BCQ_distribution.csv",index_col=0)
```

```python
df_bcq
```

```python
df_real = pd.read_csv("./consume_level_jian_buyer_download.csv",header=None)
```

```python
df_real.columns = ["buyer_id", "consume_level", "jian_level"]
```

```python
df_real.head()
```

```python
df_ref = pd.read_csv("./action_type_reference_withPos_3.csv",index_col=0)
```

```python
df_ref = df_ref.drop(columns=["Unnamed: 9"])
```

```python
df_reference = df_ref[["action_pos_type","jian_level"]]
df_reference.head()
```

```python
df_bcq_jian = df_bcq.merge(df_reference, left_on="BCQ_Action", right_on="action_pos_type")
df_bcq_jian = df_bcq_jian[["mdp_id","jian_level"]].rename(columns={"mdp_id":"buyer_id","jian_level":"BCQ_jian_level"})
df_bcq_jian = df_bcq_jian.merge(df_real, on="buyer_id")
```

```python
df_bcq_jian = df_bcq_jian[["consume_level","BCQ_jian_level","jian_level"]]
```

```python
df_bcq_jian
```

```python
bcq_dist = df_bcq_jian.groupby(["consume_level","BCQ_jian_level"]).count().reset_index().rename(columns={"jian_level":"cnt"})
```

```python
real_dist = df_bcq_jian.groupby(["consume_level","jian_level"]).count().reset_index().rename(columns={"BCQ_jian_level":"cnt"})
```

```python
bcq_dist.head()
```

```python
real_dist.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.barplot(x="consume_level",y="cnt",hue="jian_level", data=real_dist)
plt.show()
```

```python
sns.barplot(x="jian_level",y="cnt", data=real_dist[real_dist["consume_level"]==10])
plt.show()
```

```python
sns.barplot(x="jian_level",y="cnt", data=real_dist[real_dist["consume_level"]==1])
plt.show()
```

```python
sns.barplot(x="consume_level",y="cnt",hue="BCQ_jian_level", data=bcq_dist)
plt.show()
```

```python
sns.barplot(x="BCQ_jian_level",y="cnt", data=bcq_dist[real_dist["consume_level"]==10])
plt.show()
```

```python
sns.barplot(x="BCQ_jian_level",y="cnt", data=bcq_dist[real_dist["consume_level"]==1])
plt.show()
```

---

```python
# 1 and 4 jian level only
```

```python
palette = sns.color_palette(["#7428a6", "#fb98dd"])
```

```python
bcq_dist_sub = bcq_dist[bcq_dist.BCQ_jian_level.isin([1,4])]
d = {1:"LL",4:"H"}
bcq_dist_sub["jian_cat"] = bcq_dist_sub.BCQ_jian_level.apply(lambda x: d[x])
```

```python
fontsize = 18
fig, axes = plt.subplots(1,2,figsize=(23,8))
sns.barplot(x="jian_cat",y="cnt", data=bcq_dist_sub[bcq_dist_sub["consume_level"]==10],palette=palette, ax=axes[0])
sns.barplot(x="jian_cat",y="cnt", data=bcq_dist_sub[bcq_dist_sub["consume_level"]==1],palette=palette, ax=axes[1])

axes[0].set_title("High consume level",fontsize=fontsize, fontweight="bold")
axes[1].set_title("Low consume level",fontsize=fontsize, fontweight="bold")

axes[0].tick_params(labelsize=fontsize)
axes[1].tick_params(labelsize=fontsize)
axes[0].set_xlabel("Discount level", fontsize=fontsize, fontweight="bold")
axes[1].set_xlabel("Discount level", fontsize=fontsize, fontweight="bold")
axes[0].set_ylabel("Count", fontsize=fontsize, fontweight="bold")
axes[1].set_ylabel("Count", fontsize=fontsize, fontweight="bold")
plt.show()
```

```python
sns.set(font="FreeSans", style="white")
fontsize = 18
fig, ax = plt.subplots(1, figsize=(8,8))
sns.barplot(x="jian_cat",y="cnt", data=bcq_dist_sub[bcq_dist_sub["consume_level"]==10],palette=palette, ax=ax)
ax.set_title("High Spenders",fontsize=fontsize, fontweight="bold")
ax.tick_params(labelsize=fontsize)
ax.set_xlabel("Discount Level", fontsize=fontsize, fontweight="bold",fontname="FreeSans")
ax.set_ylabel("Number of Coupons", fontsize=fontsize, fontweight="bold",fontname="FreeSans")
# plt.show()
plt.tight_layout()
plt.savefig('HighSpender.pdf')
```

```python
fontsize = 18
fig, ax = plt.subplots(1, figsize=(8,8))
sns.barplot(x="jian_cat",y="cnt", data=bcq_dist_sub[bcq_dist_sub["consume_level"]==1],palette=palette, ax=ax)
ax.set_title("Low Spenders",fontsize=fontsize, fontweight="bold")
ax.tick_params(labelsize=fontsize)
ax.set_xlabel("Discount Level", fontsize=fontsize, fontweight="bold",fontname="FreeSans")
ax.set_ylabel("Number of Coupons", fontsize=fontsize, fontweight="bold",fontname="FreeSans")
# plt.show()
plt.tight_layout()
plt.savefig('LowSpender.pdf')
```

## 2020

```python
df_bcq = pd.read_csv("./BCQ_distribution_2_timeid.csv",index_col=0)
```

```python
df_bcq
```

```python
df_real = pd.read_csv("./fans_value_jian_buyer_download.csv",header=None)
```

```python
df_real.columns = ["buyer_id", "sequence_number", "fans_value"]
```

```python
df_real.head()
```

```python
df_ref = pd.read_csv("./action_type_reference_withPos_3.csv",index_col=0)
```

```python
df_ref = df_ref.drop(columns=["Unnamed: 9"])
```

```python
df_reference = df_ref[["action_pos_type","jian_level"]]
df_reference.head()
```

```python
df_bcq_jian = df_bcq.merge(df_reference, left_on="BCQ_Action", right_on="action_pos_type")
df_bcq_jian = df_bcq_jian[["mdp_id","sequence_number","jian_level"]].rename(columns={"mdp_id":"buyer_id","jian_level":"BCQ_jian_level"})
df_bcq_jian = df_bcq_jian.merge(df_real, on=["buyer_id","sequence_number"])
```

```python
df_bcq_jian = df_bcq_jian[["fans_value","BCQ_jian_level","sequence_number"]]
```

```python
df_bcq_jian
```

```python
bcq_dist = df_bcq_jian.groupby(["fans_value","BCQ_jian_level"]).count().reset_index().rename(columns={"sequence_number":"cnt"})
```

```python
# real_dist = df_bcq_jian.groupby(["consume_level","jian_level"]).count().reset_index().rename(columns={"BCQ_jian_level":"cnt"})
```

```python
bcq_dist.head()
```

```python
palette = sns.color_palette(["#7428a6", "#fb98dd"])
```

```python
bcq_dist_sub = bcq_dist[bcq_dist.BCQ_jian_level.isin([1,4])]
d = {1:"LL",4:"H"}
bcq_dist_sub["jian_cat"] = bcq_dist_sub.BCQ_jian_level.apply(lambda x: d[x])
```

```python
bcq_dist_sub
```

```python
sns.lineplot(x="fans_value",y="cnt", hue ="jian_cat", data=bcq_dist_sub)
plt.show()
```

* 对于人气的主播，BCQ会推荐更多的发H
* 对于低人气的主播，BCQ会推荐更多的发LL

```python
bcq_dist_sub[["fans_value"]].quantile(0.75)
```

```python
bcq_dist_sub["fans_value_level"] = 0
bcq_dist_sub.loc[bcq_dist_sub["fans_value"]>329,"fans_value_level"] = 1
bcq_dist_sub.loc[bcq_dist_sub["fans_value"]>537,"fans_value_level"] = 2
bcq_dist_sub.loc[bcq_dist_sub["fans_value"]>745,"fans_value_level"] = 3
```

```python
bcq_dist_sub[["fans_value"]].plot()
```

```python
bcq_dist_sub2 = bcq_dist_sub[["fans_value_level",'cnt','jian_cat']]
bcq_dist_sub2.head()
```

```python
bcq_dist_sub3 = bcq_dist_sub2.groupby(["fans_value_level","jian_cat"]).sum().reset_index()
bcq_dist_sub3.head()
```

```python
sns.barplot(x="fans_value_level",y="cnt", hue ="jian_cat", data=bcq_dist_sub)
plt.show()
```

```python
sns.barplot(x="fans_value_level",y="cnt", hue ="jian_cat", data=bcq_dist_sub2)
plt.show()
```

```python
sns.barplot(x="fans_value_level",y="cnt", hue ="jian_cat", data=bcq_dist_sub3)
plt.show()
```

```python
# sns.set_style("white")
sns.set(font="FreeSans", style="white")
```

```python
sns.set(font="FreeSans", style="white")
fontsize = 18
fig, ax = plt.subplots(1, figsize=(8,8))
sns.barplot(x="jian_cat",y="cnt", data=bcq_dist_sub3[bcq_dist_sub3["fans_value_level"]==3],palette=palette, ax=ax)
ax.set_title("Hosts with High Popularity",fontsize=fontsize, fontweight="bold")
ax.tick_params(labelsize=fontsize)
ax.set_xlabel("Discount Level", fontsize=fontsize, fontweight="bold",fontname="FreeSans")
ax.set_ylabel("Number of Coupons", fontsize=fontsize, fontweight="bold",fontname="FreeSans")
# plt.show()
plt.tight_layout()
plt.savefig('HighPopularity.pdf')
```

```python
fontsize = 18
fig, ax = plt.subplots(1, figsize=(8,8))
sns.barplot(x="jian_cat",y="cnt", data=bcq_dist_sub3[bcq_dist_sub3["fans_value_level"]==1],palette=palette, ax=ax)
ax.set_title("Hosts with Low Popularity",fontsize=fontsize, fontweight="bold")
ax.tick_params(labelsize=fontsize)
ax.set_xlabel("Discount Level", fontsize=fontsize, fontweight="bold",fontname="FreeSans")
ax.set_ylabel("Number of Coupons", fontsize=fontsize, fontweight="bold",fontname="FreeSans")
# plt.show()
plt.tight_layout()
plt.savefig('LowPopularity.pdf')
plt.show()
```
